```python
import pandas as pd
```

```python
df = pd.read_csv('movies_metadata.csv')
#df4 = df.sort_values(by=['movieId'], ascending=True).reset_index(drop=True)

df = df[df['original_language'].isin(['en'])] # film isimleri farklı alfabede yazılmış olanları dahil etmedim
df.columns
```

```python
df = df[['id','title', 'vote_average', 'vote_count', 'runtime', 'release_date', 'genres']]
df.head(n=10)
```

```python
df['genres'][0]
```

```python
df2 = pd.read_csv('ratings_small.csv')
df2 = df2.rename({'movieId': 'id'}, axis=1)
df2.head(n=10)
```

```python
df['id']=df['id'].astype(int)
result = pd.merge(df, df2, on=["id"])
result.head(n=6)
```

```python
dataset = result.sort_values(by=['userId'], ascending=True).reset_index(drop=True)
#dataset = df2[['userId','id','title','rating','vote_average','vote_count','runtime','release_date','genres','timestamp']]
#dataset = dataset[['userId','title','id','rating','vote_average','vote_count','release_date','genres']]
dataset = dataset[['userId','title','id','rating','genres']]
dataset = dataset.rename({'id': 'movieId'}, axis=1)
dataset.head(n=6)
```

```python
from ast import literal_eval
# Returns the list top l elements or entire list; whichever is more.
def get_list(x, l=5):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than l elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > l:
            names = names[:l]
        return names

    #Return empty list in case of missing/malformed data
    return []

dataset['genres'] = dataset['genres'].apply(literal_eval)
dataset['genres'] = dataset['genres'].apply(get_list)
dataset['genres']
```

```python
dataset.head(n=10)
```

```python
def classifier_movie(row):
    if  (row['genres'] == ['Drama'] or row['genres'] == ['Comedy', 'Romance'] or ['genres'] ==['Comedy', 'Drama'] 
         or ['genres'] ==['Drama', 'Romance']):
        return "emotional"
    elif  (row['genres'] == ['Comedy'] or row['genres'] == ['Comedy', 'Romance']):
        return "funny"
    elif  (row['genres'] ==['Horror', 'Mystery']  or row['genres'] ==['History'] or row['genres'] ==['Science Fiction'] 
          or row['genres'] == ['Horror'] or row['genres'] == ['Action', 'Fantasy']):
        return "curious"
    else:
        return None
dataset['type'] = dataset.apply(classifier_movie, axis=1)
dataset.head(n=10)
```

```python
dataset = dataset.dropna()
dataset = dataset.drop('genres', 1)
```
